## Problem 2 Working With Ascii Data

```python
#first we will use astropy's ascii tables to read in the output of SExtractor
r_cat=ascii.read('1608_p60_comb_r.cat')
i_cat=ascii.read('1608_p60_comb_i.cat')
```

```python
#clean up the data. Identify the r magnitudes bigger than 50
r_cat_mask=r_cat['MAG_AUTO'] > 50

#use these to identify the bad values and turn them into np.nans. save this as rmag, the cleaned version

#we have to first define rmag as an array with all zeroes with the same shape as the SExtractor magnitudes
#and then populate this array with the data from the output of SExtractor. 
#We need to do this so we avoid overwriting the data in SExtractor. previously I tried rmag=r_cat['MAG_AUTO]
#and it acted like a pointer and overwroe the original source extractor data when i was turning bad data to nans
rmag=np.zeros_like(r_cat['MAG_AUTO'].data)
rmag[:]=r_cat['MAG_AUTO'].data

#now that we have a copy of the SExtractor data, we can change bad values to nans with out worrying about 
#chaning the SExtractor data
rmag[r_cat_mask]=np.nan

#do the same for the idata
i_cat_mask=i_cat['MAG_AUTO'] > 50
imag=np.zeros_like(i_cat['MAG_AUTO'].data)
imag[:]=i_cat['MAG_AUTO'].data
imag[i_cat_mask]=np.nan

#now we make our color magnitude diagram
plt.figure(figsize=(15,8))
plt.xlabel('I', fontsize=20)
plt.ylabel('R-I', fontsize=20)
plt.plot(imag, rmag-imag, 'bo')
```

```python
#now we want to identify rdata values less than 22. we can make a mask for this, and use it to set vales
#which are 22 and higher to np.nans. We use the same process as the previous part
rmag_less22_mask=r_cat['MAG_AUTO'] >= 22
rmag_less22=np.zeros_like(r_cat['MAG_AUTO'].data)
rmag_less22[:]=r_cat['MAG_AUTO'].data
rmag_less22[rmag_less22_mask]=np.nan
```

```python
#now we may identify seperately the remaining data which has r greater than or equal to 22
rmag_greaterequal_22=np.zeros_like(r_cat['MAG_AUTO'].data)
rmag_greaterequal_22[:]=r_cat['MAG_AUTO'].data
rmag_greaterequal_22[~rmag_less22_mask]=np.nan

#we must also remember to exclude the bad data which has a magnitude greater than 50
#in the previous quesiton we identified these elements with the r_cat_mask. so we use it again to
#identify these values and set them to np.nan
rmag_greaterequal_22[r_cat_mask]=np.nan
```

```python
#now we make our color magnitude diagram
#when we do the subtraction to determine the y values, if any of the r elements is an np.nan r-i will also be
#an np.nan and ignored by matplotlib and this point will not be plotted
#likewise if any of the i values has an np.nan, the point will not be plotted
#so the blue plot only will plot points where r is greater than or equal to 22
#the red plot will only plot poits where r is less than 22
plt.figure(figsize=(15,8))
plt.xlabel('I', fontsize=20)
plt.ylabel('R-I', fontsize=20)
plt.plot(imag, rmag_greaterequal_22-imag, 'bo', label='R >= 22')
plt.plot(imag, rmag_less22 - imag, 'r^', label='R < 22')
plt.legend(loc='lower left')
```

## Problem 3 D

### There are 9 blocks in total. We do 6 draws with replacement and got a red block twice. Use Bayes Theorem to calculate the probability of Nr given the data

```python
#first define a little function which calculated P(data|Nr_i) making use of the scipy comb function
#comb computs n choose k.
from scipy.misc import comb

def posterior(Nr, M, r, n):
    '''inputs
        Nr: number of red blocks
        M: number of draws
        r: number of drawn red blocks
        n: number of blocks in the bag
    '''
    
    #calculate M choose r. use the optional input exact to calculate this exactly, instead of using an approximation
    mcx = comb(M, r, exact=True)
    
    #the next term is the product of Nr^r and (n-Nr)^(M-r)
    redAndBlue=np.power(Nr, r) * np.power(n - Nr, M-r)
    
    #note that this is un normalized
    return mcx*(redAndBlue) #/ np.power(n,M)
```

```python
#now we use this function to calculate the liklihoods of their being 0 redblocks, 1 red block and so on
#to do this lets make an array posteriors which will hold the probabilities
#so the ith element of the posteriors array will hold the probability of there being i red blocks
posteriors=np.zeros(10)

for j in xrange(0,10):
    posteriors[j] = posterior(j,6,2,9) 

#now we normalize the posteriors
posteriors = posteriors / posteriors.sum()

#now report the probabilities

for i in xrange(0,10):
    print "the liklihood of there being %d redblocks is %.7f" %(i,posteriors[i])
```

```python
#now we make a plot of our results
plt.figure(figsize=(15,8))
plt.xlabel('Nr', fontsize=20)
plt.ylabel('P(Nr)', fontsize=20)
plt.bar(np.arange(0,10), posteriors, align='center')
plt.xticks(np.arange(0,10))
```

The most likely case is that there are three red blocks, which is what we saw in part C

## Problem 3 E

### There are 9 blocks in total. We do 60 draws with replacement and got a red block 20 times. Use Bayes Theorem to calculate the probability of Nr given the data

```python
#I had to chagne the previous function a bit so it could intelligently handle large numbers
#the previous function crashed when trying to perform calculations like 8 to the power of 40

#first define a little function which calculated P(data|Nr_i) making use of the scipy comb function
#comb computs n choose k.
from scipy.misc import comb

def posterior2(Nr, M, r, n):
    '''inputs
        Nr: number of red blocks
        M: number of draws
        r: number of drawn red blocks
        n: number of blocks in the bag
    '''
    
    #calculate M choose r. use the optional input exact to calculate this exactly, instead of using an approximation
    mcx = comb(M, r, exact=True)
    
    #the next term is the product of Nr^r and (n-Nr)^(M-r)
    redAndBlue=(Nr**r) * ((n - Nr) **( M-r))
    
    #note that this is un normalized
    return mcx*(redAndBlue) #/ np.power(n,M)
```

```python
#we will follow the same procedure above, but now we will let M=60 and r=20
posteriors_60draws_20red=np.zeros(10)

for j in xrange(0,10):
    posteriors_60draws_20red[j] = posterior2(j,60,20,9) 

#now we normalize the posteriors
posteriors_60draws_20red = posteriors_60draws_20red / posteriors_60draws_20red.sum()

#now report the probabilities

for i in xrange(0,10):
    print "the liklihood of there being %d redblocks is %.7f" %(i,posteriors_60draws_20red[i])
```

```python
#now we make a plot of our results
plt.figure(figsize=(15,8))
plt.xlabel('Nr', fontsize=20)
plt.ylabel('P(Nr)', fontsize=20)
plt.bar(np.arange(0,10), posteriors_60draws_20red, align='center')
plt.xticks(np.arange(0,10))
```

With all of the new data, we have a very strong peak that suggest there are 3 red blocks. The frequentist and baysean will agree on the maximum liklihood but the baysean will be able to evaluate the Nr=3 case to all other cases
